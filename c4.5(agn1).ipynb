```python
from chefboost import Chefboost as chef
```

```python
import pandas as pd

df = pd.read_csv("mammals.csv")
```

```python
df
```

```python
config = {'algorithm': 'C4.5'}
model = chef.fit(df, config = config, target_label = 'danger')
```

```python
df
```

```python
prediction = chef.predict(model, param = ['Yellow-belliedmarmot',	4.050,	17.0,	'NaN',	'NaN',	'NaN',	13.0,	38.0,	3,	1,	1])
```
